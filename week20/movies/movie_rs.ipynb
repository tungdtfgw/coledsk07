```python
genres_example = 'Adventure|Animation|Children|Comedy|Fantasy'
genres_list = genres_example.split('|')
print(genres_list)
```

```python
movies_df['genres'] = movies_df['genres'].str.split('|')
movies_df.head()
```

```python
movies_df_encoded = movies_df.copy()
genres_cols = []

for index, row in movies_df_encoded.iterrows():
    genres_cols.append(index)
    for genre in row['genres']:
        movies_df_encoded.at[index, genre] = 1

movies_df_encoded.head()
```

```python
movies_df_encoded.fillna(0, inplace=True)
movies_df_encoded.drop(columns=['genres'], inplace=True)
movies_df_encoded = movies_df_encoded.set_index('movieId')
movies_df_encoded.head()
```

```python
ratings_df.drop(columns=['timestamp'], inplace=True)
ratings_df.head()
```

```python
tungdt_ratings = [
    {'movieId': 165, 'title': 'Die Hard: With a Vengeance (1995)', 'rating': 4},
    {'movieId': 208, 'title': 'Waterworld (1995)', 'rating': 4},
    {'movieId': 260, 'title': 'Star Wars: Episode IV - A New Hope (1977)', 'rating': 5},
    {'movieId': 5816, 'title': 'Harry Potter and the Chamber of Secrets (2002)', 'rating': 3},
    {'movieId': 8368, 'title': 'Harry Potter and the Prisoner of Azkaban (2004)', 'rating': 3},
    {'movieId': 25959, 'title': 'Annie Get Your Gun (1950)', 'rating': 2},
    {'movieId': 26590, 'title': 'G.I. Joe: The Movie (1987)', 'rating': 6},
    {'movieId': 27434, 'title': 'Darkness (2002)', 'rating': 2},
    {'movieId': 114335, 'title': 'La cravate (1957)', 'rating': 4},
    {'movieId': 130482, 'title': 'Too Late for Tears (1949)', 'rating': 3}
]
tungdt_ratings_df = pd.DataFrame(tungdt_ratings)
tungdt_ratings_df.head()
```

```python
tungdt_ids = movies_df[movies_df['movieId'].isin(tungdt_ratings_df['movieId'])]
tungdt_ids.head()
```

```python
tungdt_ratings_df = pd.merge(tungdt_ratings_df, tungdt_ids, on='movieId')
tungdt_ratings_df.head()
```

```python
tungdt_ratings_df['title'] = tungdt_ratings_df['title_x']
tungdt_ratings_df.drop(columns=['title_x', 'title_y'], inplace=True)
tungdt_ratings_df.head()
```

```python
print(movies_df_encoded.columns.tolist())
```

```python
tungdt_ratings_genres = movies_df_encoded[movies_df_encoded.title.isin(tungdt_ratings_df['title'])]
```

```python
tungdt_ratings_genres = tungdt_ratings_genres.drop(columns=['title'])
tungdt_ratings_genres.reset_index(drop=True, inplace=True)
tungdt_ratings_genres.head()
```

```python
print(tungdt_ratings_genres.shape)
```

```python
print(tungdt_ratings_df.shape)
```

```python
print(tungdt_ratings_df['rating'])
```

```python
tungdt_profile = tungdt_ratings_genres.T.dot(tungdt_ratings_df['rating'])
print(tungdt_profile)
```

```python
movies_df_encoded = movies_df_encoded.drop(columns=['title'])
movies_df_encoded.head()
```

```python
print(movies_df_encoded.shape)
```

```python
tungdt_recommendation = movies_df_encoded.dot(tungdt_profile)
tungdt_recommendation.head()
```

```python
tungdt_recommendation.sort_values(ascending=False).head(10)
```

```python
top_recommendations = tungdt_recommendation.sort_values(ascending=False).head(10)
print(top_recommendations)
```

```python
# get movie titles for the top recommendations
top_movie_ids = top_recommendations.index
top_movie_titles = movies_df[movies_df['movieId'].isin(top_movie_ids)]['title']
print(top_movie_titles)
```
