```python
def strength(duration):
    mean_duration = df_training['Duration'].mean()
    if duration >= mean_duration:
        return 'Strong'
    else:
        return 'Weak'
```

```python
mean_duration = df_training['Duration'].mean()
st = lambda duration: 'Strong' if duration >= mean_duration else 'Weak'
```

```python
print(strength(45))
```

```python
print(strength(80))
```

```python
strengths = df_training['Duration'].apply(strength)
print(strengths)
```

```python
df_training['Strength'] = strengths
df_training.head()
```

```python
# Viết tắt lại
df_training['Strength'] = df_training['Duration'].apply(strength)
df_training.head()
```
