```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
x = np.linspace(-10, 10, 100) # 100 số từ -10 đến 10
y = sigmoid(x)

def my_plot(x, y, title, xlabel, ylabel):
    plt.plot(x, y)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.show()

my_plot(x, y, 'Sigmoid Activation Function', 'x', 'sigmoid(x)')
```

```python
y = np.tanh(x)
my_plot(x, y, 'Tanh Activation Function', 'x', 'tanh(x)')
```

```python
def relu(x):
    return np.maximum(0, x)

y = relu(x)
my_plot(x, y, 'ReLU Activation Function', 'x', 'relu(x)')
```
