```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
x = np.linspace(0, 10, 100)
print(x)
y = np.sin(x)
print(y)
```

```python
def my_plot(x, y, title="My Custom Plot"):
    plt.figure(figsize=(10, 6))
    plt.plot(x, y)
    plt.title(title)
    plt.xlabel("X-axis")
    plt.ylabel("Y-axis")
    plt.grid(True)
    plt.show()
```

```python
my_plot(x, y, title="Sine Wave")
```

```python
x = np.linspace(0, 10, 20)
y = 2 * x + 1
my_plot(x, y, title="Linear Function")
```

```python
x = np.linspace(-10, 10, 100)
y = x ** 2
my_plot(x, y, title="Quadratic Function")
```

```python
products = ['Iphone 17', 'Samsung Galaxy Z', 'Google Pixel 12', 'OnePlus 11', 'Redmi Note 12']
sales = np.random.randint(1000, 5000, size=5)
list_colors = ['skyblue', 'lightgreen', 'salmon', 'orange', 'violet']
plt.figure(figsize=(10, 6))
bars = plt.bar(products, sales, color=list_colors)
plt.bar_label(bars, color='blue')
plt.title("Product Sales")
plt.xlabel("Products")
plt.ylabel("Number of Sales")
plt.show()
```

```python
plt.figure(figsize=(8, 8))
plt.pie(sales, labels=products, 
                autopct='%1.2f%%', 
                colors=list_colors, 
                startangle=140,
                explode=[0, 0.05, 0, 0, 0])
plt.title("Sales Distribution by Product")
plt.show()
```

```python
reading_scores = np.random.normal(loc=75, scale=10, size=200)
writing_scores = reading_scores + np.random.normal(loc=0, scale=5, size=200)
```

```python
print(reading_scores[:5])
print(writing_scores[:5])
```

```python
plt.scatter(reading_scores, writing_scores)
plt.title("Reading vs Writing Scores")
plt.xlabel("Reading Scores")
plt.ylabel("Writing Scores")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(reading_scores, kde=True, color='purple', bins=30)
plt.title("Distribution of Reading Scores")
plt.xlabel("Reading Scores")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(writing_scores, kde=True, color='darkorange', bins=30)
plt.title("Distribution of Writing Scores")
plt.xlabel("Writing Scores")
plt.ylabel("Frequency")
plt.show()
```

```python
df_peinguin = pd.read_csv('penguins_simple.csv')
df_peinguin.head()
```

```python
df_peinguin.info()
```

```python
df_peinguin.dropna(inplace=True)
df_peinguin.info()
```

```python
bill_length = df_peinguin['bill_length_mm']
bill_depth = df_peinguin['bill_depth_mm']
print(bill_length.shape, bill_depth.shape)
```

```python
plt.figure(figsize=(8, 8))
plt.scatter(bill_length, bill_depth, color='teal', alpha=0.7)
plt.title("Penguin Bill Length vs Bill Depth")
plt.xlabel("Bill Length (mm)")
plt.ylabel("Bill Depth (mm)")
plt.show()
```

```python
flipper_length = df_peinguin['flipper_length_mm']
plt.figure(figsize=(10, 6))
plt.scatter(bill_length, flipper_length, color='darkblue', alpha=0.4)
plt.title("Penguin Bill Length vs Flipper Length")
plt.xlabel("Bill Length (mm)")
plt.ylabel("Flipper Length (mm)")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_peinguin, x='bill_length_mm', y='bill_depth_mm', hue='species', palette='Set2', alpha=0.7)
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_peinguin, x='bill_depth_mm', y='body_mass_g', hue='species', palette='Set2', alpha=0.7)
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_peinguin, x='bill_depth_mm', y='body_mass_g', hue='island', palette='Set2', alpha=0.7)
```

```python
mean_body_mass = df_peinguin.groupby('island')['body_mass_g'].mean()
mean_body_mass.head()
```

```python
plt.bar(mean_body_mass.index, mean_body_mass.values, color=['lightblue', 'lightgreen', 'salmon'])
plt.title("Average Body Mass by Island")
plt.xlabel("Island")
plt.ylabel("Average Body Mass (g)")
plt.show()
```

```python
mean_body_mass = df_peinguin.groupby('species')['body_mass_g'].mean()
mean_body_mass.head()
plt.bar(mean_body_mass.index, mean_body_mass.values, color=['lightblue', 'lightgreen', 'salmon'])
plt.title("Average Body Mass by Species")
plt.xlabel("Species")
plt.ylabel("Average Body Mass (g)")
plt.show()
```

```python
chinstrap_female_counts = len(chinstrap_peinguins[chinstrap_peinguins['sex'] == 'female'])
chinstrap_male_counts = len(chinstrap_peinguins[chinstrap_peinguins['sex'] == 'male'])

gentoo_female_counts = len(gentoo_peinguins[gentoo_peinguins['sex'] == 'female'])
gentoo_male_counts = len(gentoo_peinguins[gentoo_peinguins['sex'] == 'male'])
```

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

axes[0].pie([adelie_male_counts, adelie_female_counts], labels=['Male', 'Female'], autopct='%1.2f%%', colors=['lightblue', 'lightpink'], startangle=0)

axes[1].pie([chinstrap_male_counts, chinstrap_female_counts], labels=['Male', 'Female'], autopct='%1.2f%%', colors=['lightblue', 'lightpink'], startangle=0)

axes[2].pie([gentoo_male_counts, gentoo_female_counts], labels=['Male', 'Female'], autopct='%1.2f%%', colors=['lightblue', 'lightpink'], startangle=0)
plt.suptitle("Gender Distribution by Species")
plt.show()
```
