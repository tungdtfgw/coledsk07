```python
a = 20

def my_func():
    a = 30      # local a
    print(a)

my_func()
print(a)    # global a
```

```python
a = 20

def my_func():
    global a
    a = 30      # global a
    print(a)

my_func()
print(a)    # global a
```

```python
def hi(name='Alice'):
    print("Hello, " + name + "!")

hi()        # sử dụng giá trị mặc định
hi("Bob")   # truyền đối số mới
```

```python
def hi(gender, name='Alice'):
    if gender == 'Male':
        print("Hello, Mr. " + name + "!")
    else:
        print("Hello, Ms. " + name + "!")
```

```python
hi('Male', 'John')
hi('Female')
```

```python
def hi(name="Alice", gender):
    if gender == 'Male':
        print("Hello, Mr. " + name + "!")
    else:
        print("Hello, Ms. " + name + "!")
```

```python
def hi(name="Alice", gender='Female'):
    if gender == 'Male':
        print("Hello, Mr. " + name + "!")
    else:
        print("Hello, Ms. " + name + "!")
```

```python
hi()
```

```python
hi(name='Ellen')
```

```python
hi(gender='Female')
```

```python
hi(name='John', gender='Male')
```

```python
hi(gender='Male', name='John')
```

```python
import random as rd
a = rd.sample(range(10, 100), 10)
print(a)
```
