```python
def area(radius):
    PI = 3.14       # constant variable
    return PI * radius * radius
```

```python
r = float(input("Enter radius: "))
a = area(r)
print(f"Area = {a:.2f}")
```

```python
def area_2(radius):
    PI = 3.14
    area = PI * radius * radius
    print(f"Area = {area:.2f}")
```

```python
r = float(input("Enter radius: "))
area_2(r)
```

```python
r1 = 5
r2 = 10
b = area(r1) + area(r2)
print(f"Total area = {b:.2f}")
a = area_2(r1) + area_2(r2)
print(f"Total area = {a:.2f}")
```

```python
def factorial(n):
    if n < 0:
        raise ValueError("Factorial is not defined for negative numbers.")
    
    if n == 0: return 1

    f = 1
    for i in range(1, n+1):
        f *= i
    return f
```

```python
print(f'0! = {factorial(0)}')
print(f'5! = {factorial(5)}')
```

```python
n = int(input('Enter n: '))
try:
    print(f'{n}! = {factorial(n)}')
except ValueError as e:
    print(e)

print('Done!')
```

```python
def is_prime(n):
    if n < 2:
        return False
    if n == 2:
        return True

    for i in range(2, n):
        if n % i == 0:
            return False
        
    return True
```

```python
print(is_prime(2))
print(is_prime(3))
print(is_prime(4))
print(is_prime(9))
```

```python
def print_primes(n):
    for i in range(2, n):
        if is_prime(i):
            print(i, end=' ')
    print()
```

```python
n = 100
print_primes(n)
```

```python
def convert(celsius):
    if celsius < -273.15:
        raise ValueError("Temperature below absolute zero is not possible.")
    
    fahrenheit = (9/5) * celsius + 32
    return fahrenheit
```

```python
c = float(input('Enter temperature in Celsius: '))
f = convert(c)
print(f'Temperature in Fahrenheit: {f:.2f}')
```
