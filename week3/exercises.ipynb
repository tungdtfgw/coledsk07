```python
nguyen_am = 'aeiou'
s = input('Enter a string: ')
print(f's = {s}')
count = 0

for c in s:
    if c in nguyen_am:
        count += 1

print(f'Số nguyên âm: {count}')
```

```python
print('Multiplication Table')
n = int(input('Enter n: '))
for i in range(1, 11):
    print(f'{n} x {i:2} = {n*i:2}')
```

```python
print_triangle = True

while print_triangle:
    n = int(input('Enter n: '))
    for i in range(1, n+1):
        for j in range(i):
            print(f'{i}', end=' ') # print i n times in the same line (no new line)
        print() # new line
    
    answer = input('Do you want to print another triangle (y/n)? ')
    print_triangle = False if answer == 'n' else True
```

```python
import random as rd

n = rd.randint(1, 10)

guessing = True

while guessing:
    user_number = int(input('Enter your guess (1-10): '))
    print(f'Your guess: {user_number}')
    if n > user_number:
        print('Too low!')
    elif n < user_number:
        print('Too high!')
    else:
        print('You got it!')
        print(f'The number is {n}')
        guessing = False
```
