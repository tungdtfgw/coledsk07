```python
fruits = ["apple", "banana", "cherry", "mango", "orange", "kiwi"]
prices = [5.99, 3.09, 4.49, 6.99, 3.49, 4.0]

m = prices[0]
m_i = 0

for i in range(1, len(prices)):
    if prices[i] > m:
        m = prices[i]
        m_i = i

print(m, m_i)
```

```python
print(fruits[m_i])
```

```python
max_price = max(prices)
max_index = prices.index(max_price)
max_fruit = fruits[max_index]
print(max_fruit, max_price)
```

```python
print(f'{sum(prices):.2f}')
```

```python
s = 0
for p in prices:
    s = s + p # s += p
print(f'{s:.2f}')
```

```python
count = 0

for i in range(len(prices)):
    if prices[i] > 4.0:
        count += 1
        print(f'{fruits[i]}: ${prices[i]:.2f}')
print(f'There are {count} fruits with price greater than $4.0')
```
