```python
a = {
    'apple': 4.5,
    'banana': 6.0,
    'cherry': 5.0,
    'mango': 6.0,
    'orange': 4.0
}
```

```python
for fruit, price in a.items():
    print(f"{fruit} : ${price}")
```

```python
max_price = max(a.values())
print(f"Max price: ${max_price}")
```

```python
for fruit, price in a.items():
    if price == max_price:
        print(f'Mosst expensive fruit: {fruit} at ${price}')
```

```python
print(a['kiwi'])
```

```python
fr = 'kiwi'
if fr in a:
    print(f'{fr}: ${a[fr]}')
else:
    print(f"{fr} not found in dictionary")
```

```python
s = 0
for fruit, price in a.items():
    s += price
print(f'Total price: ${s:.2f}')
```

```python
count = 0
for fruit, price in a.items():
    if price > 5.0:
        count += 1
        print(f'{fruit}: ${price:.2f}')

print(f'There are {count} fruits with price greater than $5.0')
```
