```python
class Human:
    def __init__(self, name):
        self.__name = name
    
    @property
    def name(self):
        return self.__name

    @name.setter
    def name(self, name):
        if not name:
            raise ValueError("Name cannot be empty.")
        self.__name = name

    def hi(self):
        print(f"Hi, my name is {self.__name}.")
```

```python
class Student(Human):
    def __init__(self, name, grade):
        super().__init__(name)  # call the constructor of the parent class (Human)
        self.grade = grade
    
    @property
    def grade(self):
        return self.__grade

    @grade.setter
    def grade(self, grade):
        if grade < 0 or grade > 10:
            raise ValueError("Grade must be between 0 and 10.")
        self.__grade = grade

    # overriding the hi method
    def hi(self):
        super().hi()  # call the hi method of the parent class
        print(f"My grade is {self.grade}.")
```

```python
john = Student("John Doe", 8)
print(john.name)  # Accessing name via getter
john.name = 'Johnny'  # Modifying name via setter
john.hi()
```

```python
print(john.grade)
john.grade = 9
print(john.grade)
john.hi()
```
