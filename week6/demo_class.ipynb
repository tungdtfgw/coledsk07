```python
class Student:
    # hàm khởi tạo (constructor)
    def __init__(self, name, age):
        # thuộc tính (attributes)
        self.name = name
        self.age = age
    # phương thức (method)
    def hi(self):
        return f"Hello, my name is {self.name} and I am {self.age} years old."
```

```python
# tạo đối tượng (object)
john = Student("John Doe", 21)
# truy cập thuộc tính của đối tượng
print(john.name)  
print(john.age)  
# gọi phương thức của đối tượng
print(john.hi()) 
```

```python
class Student:
    def __init__(self, name, age):
        self.__name = name   # private attribute
        self.age = age       # public attribute
```

```python
john = Student("John Doe", 21)
print(john.age)  # can access public attribute
print(john.__name)  # cannot access private attribute
```

```python
class Student:
    def __init__(self, name, age):
        self.__name = name
        self.__age = age

    def get_name(self):
        return self.__name

    def get_age(self):
        return self.__age
    
    def set_name(self, name):
        if name == '':
            raise ValueError("Name cannot be empty.")
        
        self.__name = name

    def set_age(self, age):
        if age < 0:
            raise ValueError("Age cannot be negative.")
        
        self.__age = age

    def hi(self):
        print(f"Hello, my name is {self.__name} and I am {self.__age} years old.")
    
    def __greet(self):
        print(f"Hi, I'm {self.__name}.")
```

```python
john = Student("John Doe", 21)
john.hi()
```

```python
print(john.get_name())
```

```python
john.set_name("Johnny")
print(john.get_name())
john.hi()
```

```python
john.set_name('')  # This will raise a ValueError
```

```python
john.set_age(25)
print(john.get_age())
john.hi()
```

```python
john.set_age(-25)  # This will raise a ValueError
```

```python
class Student:
    def __init__(self, name, age):
        self.__name = name
        self.__age = age

    def hi(self):
        print(f"Hello, my name is {self.__name} and I am {self.__age} years old.")

    @property
    def name(self):
        return self.__name

    @name.setter
    def name(self, new_name):
        if new_name == '':
            raise ValueError("Name cannot be empty.")
        self.__name = new_name

    @property
    def age(self):
        return self.__age

    @age.setter
    def age(self, new_age):
        if new_age < 0:
            raise ValueError("Age cannot be negative.")
        self.__age = new_age
```

```python
john = Student("John Doe", 21)
print(john.name)  
print(john.age)
```

```python
john.name = 'Johnny'
john.age = 25
print(john.name)  
print(john.age)
```
