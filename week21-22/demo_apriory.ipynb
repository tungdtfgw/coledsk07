```python
import pandas as pd
from apyori import apriori
```

```python
df_retail = pd.read_csv('Online_Retail_5000.csv')
df_retail.head()
```

```python
df_retail.info()
```

```python
df_retail['StockCode'].nunique()
```

```python
df_retail['InvoiceNo'].nunique()
```

```python
transactions = []
invoices = df_retail['InvoiceNo'].unique()
for i in invoices:
    items = df_retail[df_retail['InvoiceNo'] == i]['StockCode'].tolist()
    transactions.append(items)
```

```python
print(transactions[0])
```

```python
print(len(transactions))
```

```python
apr_rules = apriori(transactions, min_support=0.02, min_confidence=0.5, min_lift=1.2, max_length=5)
```

```python
apr_rules = list(apr_rules)
print(len(apr_rules))
```

```python
# sắp xếp rules theo support giảm dần
apr_rules_sorted = sorted(apr_rules, key=lambda x: x.support, reverse=True)
```

```python
# in ra 5 rules đầu tiên
for i, rule in enumerate(apr_rules_sorted[:5]):
    print(f'Rule {i+1}:')
    print(f'Items: {list(rule.items)}')
    print(f'Support: {rule.support}')
    #print(f'Ordered Statistics: {rule.ordered_statistics}')
    for stat in rule.ordered_statistics:
        print(f'Rule {list(stat.items_base)} -> {list(stat.items_add)}')
        print(f'Confidence: {stat.confidence}')
        print(f'Lift: {stat.lift}')
    print('-------------------------')
```
