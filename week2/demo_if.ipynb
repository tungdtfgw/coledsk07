```python
age = int(input('Nhập tuổi: '))
print(f'Tuổi của bạn là: {age}')
if age >= 18:
    print('Bạn đủ tuổi bầu cử')
else:
    print('Bạn chưa đủ tuổi bầu cử')
```

```python
grade = int(input('Nhâp điểm của bạn: '))

if grade <= 5:
    print('Trung bình')
elif grade < 8:
    print('Khá')
else:
    print('Giỏi')
```
