```python
width = float(input('Enter width: '))
height = float(input('Enter height: '))

if width > 0 and height > 0: # kiểm tra điều kiện đúng truớc khi tính toán
    area = width * height
    perimeter = 2 * (width + height)
    print(f'Perimeter = {perimeter}')
    print(f'Area = {area}')
else:
    print('Error: Width and height must be positive numbers.')
```

```python
# Bài tập: tính chỉ số BMI theo chiều cao và cân nặng người dùng nhập vào
# Dưới 18.5: Thiếu cân/Gầy.
# 18.5 - 22.9: Bình thường/Cân đối.
# 23 - 24.9: Thừa cân.
# Từ 25 trở lên: Béo phì. 
```

```python
height = float(input('Nhập chiều cao (m): '))
weight = float(input('Nhập cân nặng (kg): '))

if height < 1 or height > 2:
    print('Chiều cao không hợp lệ!')
elif weight < 10 or weight > 200:
    print('Cân nặng không hợp lệ!')
else:
    bmi = weight / (height ** 2)
    print(f'Chỉ số BMI của bạn là: {bmi:.2f}')
    
    if bmi < 18.5:
        print('Bạn thiếu cân')
    elif bmi < 23:
        print('Bạn có cân nặng bình thường')
    elif bmi < 25:
        print('Bạn thừa cân')
    else:
        print('Bạn bị béo phì')
```

```python
month = int(input('Nhập tháng (1-12): '))
if month < 1 or month > 12:
    print('Tháng không hợp lệ!')
else:
    if month == 2:
        print(f'Tháng {month} có 28 hoặc 29 ngày')
    elif month in [1, 3, 5, 7, 8, 10, 12]:
        print(f'Tháng {month} có 31 ngày')
    else:
        print(f'Tháng {month} có 30 ngày')
```
