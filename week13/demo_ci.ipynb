```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

1 công ty có 50 nhân viên, lương trung bình 18K usd, độ lệch chuẩn 3K. Tính khoảng tin cậy 95% cho lương trung bình

```python
n_employees = 50
std_salary = 3
mean_salary = 18

salaries = np.random.normal(size=n_employees, loc=mean_salary, scale=std_salary)

print(np.mean(salaries))
print(np.std(salaries))
```

```python
plt.hist(salaries, bins=10, color='cyan', edgecolor='black')
plt.show()
```

```python
confidence = 0.95
alpha = 1 - confidence # 0.05

z = stats.norm.ppf(1 - alpha/2)
print(z)
```

```python
margin = z * (std_salary / np.sqrt(n_employees))
print(margin)

lower_bound = mean_salary - margin
upper_bound = mean_salary + margin

print(f"{confidence}% CI for mean salary: {lower_bound:.2f} - {upper_bound:.2f}")
```

Khảo sát 20 công nhân, thời gian trung bình hoàn thành sp là 10p, độ lệch chuẩn là 2p.

```python
n_workers = 20
time_mean = 10
time_std = 2

confidence = 0.95
alpha = 1 - confidence

t = stats.t.ppf(1 - alpha/2, df=n_workers-1)
print(t)
```

```python
margin = t * (time_std / np.sqrt(n_workers))
print(margin)
```

```python
lower_bound = time_mean - margin
upper_bound = time_mean + margin

print(f"95% CI in : {lower_bound} - {upper_bound}")
```

Khảo sát 200 khách hàng, 160 người là hài lòng. Tính khoảng tin cậy 99%

```python
n_customers = 200
n_happy = 160
p = n_happy / n_customers
print(p)
```

```python
confidence = 0.99
alpha = 1 - confidence

z = stats.norm.ppf(1 - alpha/2)
print(z)
```

```python
margin = z * ((p * (1 - p)) / np.sqrt(n_customers))
lower = p - margin
upper = p + margin
print(f"{confidence * 100}% tỉ lệ hài lòng trong khoảng {lower:.2f} - {upper:.2f}")
```

Trường học công bố điểm trung bình môn Toán là 7.5 Làm thống kê để kiểm tra xem điểm TB có đúng là 7.5 với mức ý nghĩa alpha = 0.01

```python
mean_declare = 7.5
# H0: mean = 7.5
# H1: mean != 7.5
n_students = 25
```

```python
scores = np.random.normal(size=n_students, loc=7.4, scale=1)
mean_sample = np.mean(scores)
std_sample = np.std(scores)

print(mean_sample, std_sample)
```

```python
alpha = 0.01
t = (mean_sample - mean_declare) / (std_sample / np.sqrt(n_students))
p_value = 2 * (1 - stats.t.cdf(abs(t), df=n_students-1))
print(t, p_value)
```

```python
if p_value <= alpha:
  print('Bác bỏ giả thuyết H0')
else:
  print('Không bác bỏ giả thuyết H0')
```

Nhà máy sản xuất 1 sản phẩm, trọng lượng đăng ký chuẩn là 500g, độ lệch chuẩn là 3g.
H0: trọng lượng trung bình < 500g
Lấy mẫu 100 sản phẩm để kiểm định

```python
mean_declare = 500
std_weight = 3
n_samples = 100
```

```python
weights = np.random.normal(size=n_samples, scale=std_weight, loc=600)
mean_sample = np.mean(weights)
std_sample = np.std(weights)
print(mean_sample, std_sample)
```

```python
z = (mean_sample - mean_declare) / (std_sample / np.sqrt(n_samples))
print(z)
p_value = 1 - stats.norm.cdf(z)

alpha = 0.05

if p_value <= alpha:
  print('Bác bỏ giả thuyết H0')
else:
  print('Không bác bỏ giả thuyết H0')
```
