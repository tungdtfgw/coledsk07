```python
import pandas as pd
import numpy as np
```

```python
df_diemthi = pd.DataFrame({
    'Grade': np.random.randint(50, 100, size=500)
})
```

```python
df_diemthi.head()
```

```python
# Lấy mâu ngẫu nhiên 50 dòng từ DataFrame, random_state để đảm bảo tính tái lập
df_sample = df_diemthi.sample(n=50, random_state=42)
df_sample.head()
```

```python
# Thêm cột giới tính với giá trị ngẫu nhiên
df_diemthi['Gender'] = np.random.choice(['Male', 'Female'], size=len(df_diemthi))
df_diemthi.head()
```

```python
n_males = len(df_diemthi[df_diemthi['Gender'] == 'Male'])
n_females = len(df_diemthi[df_diemthi['Gender'] == 'Female'])
print(f"Số lượng nam trong mẫu: {n_males}")
print(f"Số lượng nữ trong mẫu: {n_females}")
```

```python
# Lấy mẫu ngẫu nhiên 50 dòng từ Data
df_sample = df_diemthi.sample(n=50)
n_males = len(df_sample[df_sample['Gender'] == 'Male'])
n_females = len(df_sample[df_sample['Gender'] == 'Female'])
print(f"Số lượng nam trong mẫu: {n_males}")
print(f"Số lượng nữ trong mẫu: {n_females}")
```

```python
def random_sample(df, n_samples):
    return df.sample(n=n_samples)

df_groups = df_diemthi.groupby('Gender').apply(random_sample, n_samples=25)
# df_groups = df_diemthi.groupby('Gender').apply(lambda x: x.sample(n=25))
n_males = len(df_groups[df_groups['Gender'] == 'Male'])
n_females = len(df_groups[df_groups['Gender'] == 'Female'])
print(f"Số lượng nam trong mẫu nhóm: {n_males}")
print(f"Số lượng nữ trong mẫu nhóm: {n_females}")
```
